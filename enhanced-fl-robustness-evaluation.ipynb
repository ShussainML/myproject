```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Subset, Dataset
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import copy
import time
import json
import warnings
from collections import defaultdict
from datetime import datetime
from pathlib import Path
from sklearn.metrics import classification_report, confusion_matrix
import scipy.stats as stats
warnings.filterwarnings('ignore')

class EnhancedExperimentConfig:
    """Optimized configuration for 5-7 hour Kaggle execution"""
    def __init__(self):
        # Balanced settings for thorough yet fast evaluation
        self.num_clients = 12          # Reasonable federation size
        self.clients_per_round = 6     # 50% participation  
        self.num_rounds = 4            # Sufficient for convergence
        self.local_epochs = 1          # Fast local training
        self.batch_size = 128          # Efficient batch size
        self.learning_rate = 0.05      # Fast convergence
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        
        # Data settings
        self.alpha_dirichlet = 0.5
        self.min_samples_per_client = 200
        
        # Experiment settings
        self.num_runs = 1              # Single run for speed
        self.seed_base = 42
        self.save_results = True
        
        # Enhanced method selection - 3 core methods
        self.fl_methods = [
            'FedAvg',     # Baseline
            'FedProx',    # Handles non-IID
            'FedNova',    # Handles unequal work (NEW!)
        ]
        
        # Multi-dataset support - 3 datasets
        self.datasets = ['MNIST', 'Fashion-MNIST', 'CIFAR-10']
        
        # Attack scenarios per dataset (5 each)
        self.attack_scenarios = {
            'MNIST': [
                'Clean',
                'Byzantine_20',
                'DLG_Privacy',
                'Client_Dropout_30',
                'Extreme_NonIID',
            ],
            'Fashion-MNIST': [
                'Clean',
                'Byzantine_20', 
                'DLG_Privacy',
                'Client_Dropout_30',
                'Extreme_NonIID',
            ],
            'CIFAR-10': [
                'Clean',
                'Byzantine_20',
                'DLG_Privacy', 
                'Client_Dropout_30',
                'Extreme_NonIID',
            ]
        }
        
        # Calculate total experiments
        total_experiments = sum(len(scenarios) * len(self.fl_methods) * self.num_runs 
                              for scenarios in self.attack_scenarios.values())
        
        print(f"ðŸš€ ENHANCED FL ROBUSTNESS EVALUATION:")
        print(f"   Target: Complete in 5-7 hours")
        print(f"   Datasets: {len(self.datasets)} - {', '.join(self.datasets)}")
        print(f"   Methods: {len(self.fl_methods)} - {', '.join(self.fl_methods)}")
        print(f"   Attack Scenarios: {sum(len(s) for s in self.attack_scenarios.values())}")
        print(f"   Rounds per experiment: {self.num_rounds}")
        print(f"   Local epochs: {self.local_epochs}")
        print(f"   Total experiments: {total_experiments}")
        print(f"   Estimated time: ~6 hours (8min per experiment)")
        print(f"   Execution order: Attack-by-attack across all methods & datasets")

class AdaptiveModel(nn.Module):
    """Lightweight model that adapts to different datasets"""
    def __init__(self, input_channels=1, input_size=28, num_classes=10):
        super().__init__()
        self.input_channels = input_channels
        self.input_size = input_size
        
        # Adaptive first layer based on input
        if input_size == 28:  # MNIST, Fashion-MNIST
            self.conv1 = nn.Conv2d(input_channels, 32, 5)
            self.pool = nn.MaxPool2d(2, 2)
            self.conv2 = nn.Conv2d(32, 64, 5)
            self.fc1_input = 64 * 4 * 4  # After two conv+pool layers
        else:  # CIFAR-10 (32x32)
            self.conv1 = nn.Conv2d(input_channels, 32, 5)
            self.pool = nn.MaxPool2d(2, 2)
            self.conv2 = nn.Conv2d(32, 64, 5)
            self.fc1_input = 64 * 5 * 5  # After two conv+pool layers
            
        self.fc1 = nn.Linear(self.fc1_input, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, num_classes)
        self.dropout = nn.Dropout(0.25)
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

def create_non_iid_data(dataset, num_clients, alpha=0.5, min_samples=200, num_classes=10):
    """Enhanced non-IID data creation"""
    class_indices = {i: [] for i in range(num_classes)}
    
    # Collect indices by class
    for idx, (_, label) in enumerate(dataset):
        if label < num_classes:
            class_indices[label].append(idx)
    
    client_indices = [[] for _ in range(num_clients)]
    
    # Dirichlet distribution for realistic non-IID split
    for class_id in range(num_classes):
        indices = class_indices[class_id]
        if len(indices) == 0:
            continue
            
        np.random.shuffle(indices)
        proportions = np.random.dirichlet(np.repeat(alpha, num_clients))
        split_points = (np.cumsum(proportions) * len(indices)).astype(int)[:-1]
        splits = np.split(indices, split_points)
        
        for client_id, split in enumerate(splits):
            if client_id < num_clients:
                client_indices[client_id].extend(split)
    
    # Ensure minimum samples per client
    valid_clients = []
    for indices in client_indices:
        if len(indices) >= min_samples:
            valid_clients.append(indices)
        elif len(indices) > 0:  # Redistribute small clients
            valid_clients.append(indices)
    
    return valid_clients[:num_clients]

# ================================
# MULTI-DATASET LOADERS
# ================================

def load_mnist(config, attack_scenario='Clean'):
    """Load MNIST dataset"""
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
    ])
    
    train_dataset = torchvision.datasets.MNIST(root='./data', train=True,
                                             download=True, transform=transform)
    test_dataset = torchvision.datasets.MNIST(root='./data', train=False,
                                            download=True, transform=transform)
    
    alpha = config.alpha_dirichlet
    if 'Extreme_NonIID' in attack_scenario:
        alpha = 0.1
    
    client_indices = create_non_iid_data(
        train_dataset, config.num_clients, alpha=alpha,
        min_samples=config.min_samples_per_client, num_classes=10
    )
    
    client_loaders = []
    for indices in client_indices:
        subset = Subset(train_dataset, indices)
        loader = DataLoader(subset, batch_size=config.batch_size, 
                          shuffle=True, num_workers=0, pin_memory=True)
        client_loaders.append(loader)
    
    test_loader = DataLoader(test_dataset, batch_size=200, shuffle=False,
                           num_workers=0, pin_memory=True)
    
    return client_loaders, test_loader

def load_fashion_mnist(config, attack_scenario='Clean'):
    """Load Fashion-MNIST dataset"""
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.2860,), (0.3530,))
    ])
    
    train_dataset = torchvision.datasets.FashionMNIST(root='./data', train=True,
                                                    download=True, transform=transform)
    test_dataset = torchvision.datasets.FashionMNIST(root='./data', train=False,
                                                   download=True, transform=transform)
    
    alpha = config.alpha_dirichlet
    if 'Extreme_NonIID' in attack_scenario:
        alpha = 0.1
    
    client_indices = create_non_iid_data(
        train_dataset, config.num_clients, alpha=alpha,
        min_samples=config.min_samples_per_client, num_classes=10
    )
    
    client_loaders = []
    for indices in client_indices:
        subset = Subset(train_dataset, indices)
        loader = DataLoader(subset, batch_size=config.batch_size,
                          shuffle=True, num_workers=0, pin_memory=True)
        client_loaders.append(loader)
    
    test_loader = DataLoader(test_dataset, batch_size=200, shuffle=False,
                           num_workers=0, pin_memory=True)
    
    return client_loaders, test_loader

def load_cifar10(config, attack_scenario='Clean'):
    """Load CIFAR-10 dataset"""
    transform_train = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    ])
    
    transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    ])
    
    train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                               download=True, transform=transform_train)
    test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                              download=True, transform=transform_test)
    
    alpha = config.alpha_dirichlet
    if 'Extreme_NonIID' in attack_scenario:
        alpha = 0.1
    
    client_indices = create_non_iid_data(
        train_dataset, config.num_clients, alpha=alpha,
        min_samples=config.min_samples_per_client, num_classes=10
    )
    
    client_loaders = []
    for indices in client_indices:
        subset = Subset(train_dataset, indices)
        loader = DataLoader(subset, batch_size=config.batch_size,
                          shuffle=True, num_workers=0, pin_memory=True)
        client_loaders.append(loader)
    
    test_loader = DataLoader(test_dataset, batch_size=200, shuffle=False,
                           num_workers=0, pin_memory=True)
    
    return client_loaders, test_loader

def load_dataset(dataset_name, config, attack_scenario='Clean'):
    """Universal dataset loader"""
    loaders = {
        'MNIST': load_mnist,
        'Fashion-MNIST': load_fashion_mnist,
        'CIFAR-10': load_cifar10,
    }
    return loaders[dataset_name](config, attack_scenario)

# ================================
# ENHANCED FL METHODS
# ================================

class EnhancedFedAvgServer:
    """Enhanced FedAvg implementation"""
    def __init__(self, config, dataset_info):
        self.config = config
        self.dataset_info = dataset_info
        self.model = AdaptiveModel(
            input_channels=dataset_info['channels'],
            input_size=dataset_info['size'],
            num_classes=dataset_info['classes']
        ).to(config.device)
        self.metrics = {'accuracy': [], 'loss': [], 'round': []}
        self.method_name = "FedAvg"
        
    def aggregate(self, client_models, client_sizes, **kwargs):
        """Weighted averaging aggregation"""
        total_size = sum(client_sizes)
        weights = [size / total_size for size in client_sizes]
        
        global_state = self.model.state_dict()
        
        for key in global_state.keys():
            global_state[key].zero_()
            for model, weight in zip(client_models, weights):
                global_state[key] += weight * model.state_dict()[key]
        
        self.model.load_state_dict(global_state)
        return copy.deepcopy(self.model)
    
    def evaluate(self, test_loader):
        """Model evaluation"""
        self.model.eval()
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in test_loader:
                data, target = data.to(self.config.device), target.to(self.config.device)
                output = self.model(data)
                pred = output.argmax(dim=1)
                correct += pred.eq(target).sum().item()
                total += target.size(0)
        
        accuracy = 100. * correct / total
        self.metrics['accuracy'].append(accuracy)
        self.metrics['round'].append(len(self.metrics['round']) + 1)
        
        return accuracy, 0.0

class EnhancedFedProxServer:
    """Enhanced FedProx implementation"""
    def __init__(self, config, dataset_info):
        self.config = config
        self.dataset_info = dataset_info
        self.model = AdaptiveModel(
            input_channels=dataset_info['channels'],
            input_size=dataset_info['size'],
            num_classes=dataset_info['classes']
        ).to(config.device)
        self.metrics = {'accuracy': [], 'loss': [], 'round': []}
        self.method_name = "FedProx"
        self.mu = 0.01  # Proximal term coefficient
        
    def aggregate(self, client_models, client_sizes, **kwargs):
        """Same as FedAvg - difference is in client training"""
        total_size = sum(client_sizes)
        weights = [size / total_size for size in client_sizes]
        
        global_state = self.model.state_dict()
        
        for key in global_state.keys():
            global_state[key].zero_()
            for model, weight in zip(client_models, weights):
                global_state[key] += weight * model.state_dict()[key]
        
        self.model.load_state_dict(global_state)
        return copy.deepcopy(self.model)
    
    def evaluate(self, test_loader):
        """Model evaluation"""
        self.model.eval()
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in test_loader:
                data, target = data.to(self.config.device), target.to(self.config.device)
                output = self.model(data)
                pred = output.argmax(dim=1)
                correct += pred.eq(target).sum().item()
                total += target.size(0)
        
        accuracy = 100. * correct / total
        self.metrics['accuracy'].append(accuracy)
        self.metrics['round'].append(len(self.metrics['round']) + 1)
        
        return accuracy, 0.0

class FedNovaServer:
    """NEW! FedNova (Normalized Averaging) implementation"""
    def __init__(self, config, dataset_info):
        self.config = config
        self.dataset_info = dataset_info
        self.model = AdaptiveModel(
            input_channels=dataset_info['channels'],
            input_size=dataset_info['size'],
            num_classes=dataset_info['classes']
        ).to(config.device)
        self.metrics = {'accuracy': [], 'loss': [], 'round': []}
        self.method_name = "FedNova"
        
    def aggregate(self, client_models, client_sizes, client_steps=None, **kwargs):
        """FedNova normalized averaging"""
        if client_steps is None:
            client_steps = [1] * len(client_models)  # Default to 1 step per client
        
        # Calculate effective learning rates (normalized by steps)
        total_size = sum(client_sizes)
        
        # FedNova normalization
        normalized_weights = []
        total_norm = 0
        
        for size, steps in zip(client_sizes, client_steps):
            # Normalize by both data size and number of local steps
            norm_weight = size / total_size
            if steps > 0:
                norm_weight = norm_weight / steps
            normalized_weights.append(norm_weight)
            total_norm += norm_weight
        
        # Renormalize
        if total_norm > 0:
            normalized_weights = [w / total_norm for w in normalized_weights]
        else:
            normalized_weights = [1.0 / len(client_models)] * len(client_models)
        
        # Aggregate with normalized weights
        global_state = self.model.state_dict()
        
        for key in global_state.keys():
            global_state[key].zero_()
            for model, weight in zip(client_models, normalized_weights):
                global_state[key] += weight * model.state_dict()[key]
        
        self.model.load_state_dict(global_state)
        return copy.deepcopy(self.model)
    
    def evaluate(self, test_loader):
        """Model evaluation"""
        self.model.eval()
        correct = 0
        total = 0
        
        with torch.no_grad():
            for data, target in test_loader:
                data, target = data.to(self.config.device), target.to(self.config.device)
                output = self.model(data)
                pred = output.argmax(dim=1)
                correct += pred.eq(target).sum().item()
                total += target.size(0)
        
        accuracy = 100. * correct / total
        self.metrics['accuracy'].append(accuracy)
        self.metrics['round'].append(len(self.metrics['round']) + 1)
        
        return accuracy, 0.0

# ================================
# ENHANCED CLIENTS
# ================================

class EnhancedStandardClient:
    """Enhanced standard client with step tracking"""
    def __init__(self, client_id, data_loader, config):
        self.client_id = client_id
        self.data_loader = data_loader
        self.config = config
        self.local_steps = 0
        
    def train(self, global_model, method_name="FedAvg", **kwargs):
        """Enhanced local training with step tracking"""
        model = copy.deepcopy(global_model)
        model.train()
        
        optimizer = optim.SGD(model.parameters(), lr=self.config.learning_rate, 
                            momentum=0.9, weight_decay=1e-4)
        
        # Store global params for FedProx
        if method_name == "FedProx":
            global_params = {name: param.clone() for name, param in global_model.named_parameters()}
            mu = 0.01
        
        self.local_steps = 0
        
        # Local training epochs
        for epoch in range(self.config.local_epochs):
            for batch_idx, (data, target) in enumerate(self.data_loader):
                data, target = data.to(self.config.device), target.to(self.config.device)
                
                optimizer.zero_grad()
                output = model(data)
                loss = F.cross_entropy(output, target)
                
                # Add FedProx regularization
                if method_name == "FedProx":
                    prox_term = 0.0
                    for name, param in model.named_parameters():
                        prox_term += (mu / 2) * torch.norm(param - global_params[name]) ** 2
                    loss += prox_term
                
                loss.backward()
                optimizer.step()
                self.local_steps += 1
        
        return model

class EnhancedByzantineClient:
    """Enhanced Byzantine client"""
    def __init__(self, client_id, data_loader, config):
        self.client_id = client_id
        self.data_loader = data_loader
        self.config = config
        self.local_steps = 0
        
    def train(self, global_model, method_name="FedAvg", **kwargs):
        """Enhanced malicious training"""
        model = copy.deepcopy(global_model)
        
        # More sophisticated attack: Gaussian noise + sign flipping
        with torch.no_grad():
            for param in model.parameters():
                noise = torch.randn_like(param.data) * 0.1
                param.data = -param.data + noise
        
        self.local_steps = 1
        return model

# ================================
# ENHANCED ATTACK IMPLEMENTATIONS
# ================================

class EnhancedDLGAttack:
    """Enhanced DLG (Deep Leakage from Gradients) attack"""
    def __init__(self, model, device):
        self.model = model
        self.device = device
        
    def attack(self, gradients, original_data, original_labels, num_iterations=15):
        """Enhanced DLG attack with better optimization"""
        dummy_data = torch.randn(original_data.shape, requires_grad=True, device=self.device)
        dummy_labels = torch.randint(0, 10, original_labels.shape, device=self.device)
        
        optimizer = optim.LBFGS([dummy_data], lr=0.1)
        
        for iteration in range(num_iterations):
            def closure():
                optimizer.zero_grad()
                self.model.zero_grad()
                
                output = self.model(dummy_data)
                loss = F.cross_entropy(output, dummy_labels)
                dummy_gradients = torch.autograd.grad(loss, self.model.parameters(), create_graph=True)
                
                grad_loss = sum((grad_dummy - grad_real).pow(2).sum() 
                              for grad_dummy, grad_real in zip(dummy_gradients, gradients))
                grad_loss.backward()
                
                return grad_loss
            
            optimizer.step(closure)
        
        mse = F.mse_loss(dummy_data, original_data).item()
        return mse, dummy_data

def simulate_client_dropout(client_indices, dropout_rate=0.3):
    """Enhanced client dropout simulation"""
    num_clients = len(client_indices)
    num_dropout = int(num_clients * dropout_rate)
    
    if num_dropout > 0:
        dropout_indices = np.random.choice(num_clients, num_dropout, replace=False)
        remaining_indices = [i for i in range(num_clients) if i not in dropout_indices]
        return [client_indices[i] for i in remaining_indices]
    
    return client_indices

# ================================
# ENHANCED EXPERIMENT RUNNER
# ================================

class EnhancedRobustnessEvaluator:
    """Enhanced experiment runner with attack-by-attack execution"""
    def __init__(self, config):
        self.config = config
        self.results = defaultdict(list)
        
        # Dataset info
        self.dataset_info = {
            'MNIST': {'channels': 1, 'size': 28, 'classes': 10},
            'Fashion-MNIST': {'channels': 1, 'size': 28, 'classes': 10},
            'CIFAR-10': {'channels': 3, 'size': 32, 'classes': 10},
        }
        
    def create_server(self, method_name, dataset_name):
        """Enhanced server creation"""
        dataset_info = self.dataset_info[dataset_name]
        servers = {
            "FedAvg": EnhancedFedAvgServer,
            "FedProx": EnhancedFedProxServer,
            "FedNova": FedNovaServer,  # NEW!
        }
        return servers[method_name](self.config, dataset_info)
    
    def create_clients(self, client_loaders, attack_scenario):
        """Enhanced client creation"""
        clients = []
        
        # Determine attack parameters
        byzantine_ratio = 0.2 if 'Byzantine_20' in attack_scenario else 0.0
        num_malicious = int(len(client_loaders) * byzantine_ratio)
        
        for i, loader in enumerate(client_loaders):
            if i < num_malicious:
                clients.append(EnhancedByzantineClient(i, loader, self.config))
            else:
                clients.append(EnhancedStandardClient(i, loader, self.config))
        
        return clients
    
    def run_single_experiment(self, method_name, dataset_name, attack_scenario, run_id):
        """Enhanced single experiment"""
        print(f"  ðŸ”¬ {method_name} on {dataset_name} vs {attack_scenario}")
        
        # Set seed
        torch.manual_seed(self.config.seed_base + run_id)
        np.random.seed(self.config.seed_base + run_id)
        
        # Load data
        client_loaders, test_loader = load_dataset(dataset_name, self.config, attack_scenario)
        
        # Create server and clients
        server = self.create_server(method_name, dataset_name)
        clients = self.create_clients(client_loaders, attack_scenario)
        client_sizes = [len(loader.dataset) for loader in client_loaders]
        
        # Training loop
        attack_metrics = {'dlg_mse': [], 'reconstruction_success': []}
        
        for round_num in range(self.config.num_rounds):
            # Client selection
            available_clients = list(range(len(clients)))
            
            if 'Client_Dropout_30' in attack_scenario:
                available_clients = simulate_client_dropout(available_clients, dropout_rate=0.3)
            
            # Select clients
            if len(available_clients) >= self.config.clients_per_round:
                selected_indices = np.random.choice(
                    available_clients, self.config.clients_per_round, replace=False)
            else:
                selected_indices = available_clients
            
            selected_clients = [clients[i] for i in selected_indices]
            selected_sizes = [client_sizes[i] for i in selected_indices]
            
            # Local training
            client_models = []
            client_steps = []
            
            for client in selected_clients:
                model = client.train(server.model, method_name)
                client_models.append(model)
                client_steps.append(client.local_steps)
            
            # Privacy attack (DLG) - first round only
            if attack_scenario == 'DLG_Privacy' and round_num == 0 and len(client_models) > 0:
                try:
                    sample_data, sample_labels = next(iter(client_loaders[selected_indices[0]]))
                    sample_data = sample_data[:1].to(self.config.device)
                    sample_labels = sample_labels[:1].to(self.config.device)
                    
                    client_models[0].eval()
                    output = client_models[0](sample_data)
                    loss = F.cross_entropy(output, sample_labels)
                    gradients = torch.autograd.grad(loss, client_models[0].parameters())
                    
                    dlg_attack = EnhancedDLGAttack(client_models[0], self.config.device)
                    mse, _ = dlg_attack.attack(gradients, sample_data, sample_labels)
                    
                    attack_metrics['dlg_mse'].append(mse)
                    attack_metrics['reconstruction_success'].append(mse < 1.0)
                except:
                    pass  # Skip on error
            
            # Server aggregation (pass client steps for FedNova)
            server.aggregate(client_models, selected_sizes, client_steps=client_steps)
            
            # Evaluation
            accuracy, _ = server.evaluate(test_loader)
        
        # Collect results
        result = {
            'method': method_name,
            'dataset': dataset_name,
            'attack': attack_scenario,
            'run_id': run_id,
            'final_accuracy': server.metrics['accuracy'][-1] if server.metrics['accuracy'] else 0.0,
            'best_accuracy': max(server.metrics['accuracy']) if server.metrics['accuracy'] else 0.0,
            'accuracy_history': server.metrics['accuracy'],
            'attack_metrics': attack_metrics
        }
        
        return result
    
    def run_all_experiments(self):
        """Run experiments attack-by-attack for better comparison"""
        all_attacks = set()
        for attacks in self.config.attack_scenarios.values():
            all_attacks.update(attacks)
        all_attacks = sorted(list(all_attacks))
        
        total_experiments = sum(len(scenarios) * len(self.config.fl_methods) * self.config.num_runs 
                              for scenarios in self.config.attack_scenarios.values())
        experiment_count = 0
        
        print(f"ðŸš€ Starting ENHANCED evaluation: {total_experiments} experiments")
        print(f"ðŸ“‹ Execution order: Attack-by-attack across all methods & datasets")
        start_time = time.time()
        
        # Execute attack-by-attack
        for attack_type in all_attacks:
            print(f"\nðŸŽ¯ EVALUATING ATTACK: {attack_type}")
            print("=" * 60)
            
            attack_start_time = time.time()
            attack_experiments = 0
            
            for dataset_name in self.config.datasets:
                if attack_type in self.config.attack_scenarios[dataset_name]:
                    for method in self.config.fl_methods:
                        for run_id in range(self.config.num_runs):
                            experiment_count += 1
                            attack_experiments += 1
                            
                            elapsed_time = time.time() - start_time
                            estimated_total = elapsed_time * total_experiments / experiment_count if experiment_count > 0 else 0
                            remaining_time = estimated_total - elapsed_time
                            
                            print(f"\nðŸ“Š Progress: {experiment_count}/{total_experiments}")
                            print(f"   Elapsed: {elapsed_time/3600:.1f}h, Est. Remaining: {remaining_time/3600:.1f}h")
                            
                            try:
                                result = self.run_single_experiment(method, dataset_name, attack_type, run_id)
                                self.results[f"{dataset_name}_{method}_{attack_type}"].append(result)
                                print(f"    âœ… {result['final_accuracy']:.1f}% accuracy")
                                
                            except Exception as e:
                                print(f"    âŒ Failed: {e}")
                                continue
            
            attack_time = time.time() - attack_start_time
            print(f"\nâœ… Attack {attack_type} completed: {attack_experiments} experiments in {attack_time/60:.1f} minutes")
        
        total_time = time.time() - start_time
        print(f"\nðŸŽ‰ ALL EXPERIMENTS COMPLETED!")
        print(f"â±ï¸  Total time: {total_time/3600:.1f} hours")
        print(f"ðŸ“Š Total experiments: {experiment_count}")
        print(f"âš¡ Average time per experiment: {total_time/experiment_count/60:.1f} minutes")
        
        return self.results

# ================================
# ENHANCED ANALYSIS
# ================================

class EnhancedResultsAnalyzer:
    """Enhanced results analysis with multi-dataset support"""
    def __init__(self, results):
        self.results = results
        self.df = self._create_dataframe()
        
    def _create_dataframe(self):
        """Convert to DataFrame"""
        rows = []
        for key, experiments in self.results.items():
            for exp in experiments:
                rows.append({
                    'method': exp['method'],
                    'dataset': exp.get('dataset', 'Unknown'),
                    'attack': exp['attack'],
                    'run_id': exp['run_id'],
                    'final_accuracy': exp['final_accuracy'],
                    'best_accuracy': exp['best_accuracy']
                })
        return pd.DataFrame(rows)
    
    def comprehensive_analysis(self):
        """Comprehensive analysis and visualization"""
        print("\nðŸ“Š COMPREHENSIVE RESULTS ANALYSIS")
        print("=" * 60)
        
        if self.df.empty:
            print("No results to analyze!")
            return
        
        # Overall summary
        print("\nðŸ“ˆ Overall Performance Summary:")
        overall_summary = self.df.groupby(['method', 'dataset', 'attack'])['final_accuracy'].agg(['mean', 'std']).round(1)
        print(overall_summary)
        
        # Method comparison across datasets
        print("\nðŸ”¬ Method Comparison Across Datasets:")
        method_comparison = self.df.groupby(['method', 'dataset'])['final_accuracy'].mean().unstack()
        print(method_comparison.round(1))
        
        # Attack impact analysis
        print("\nðŸŽ¯ Attack Impact Analysis:")
        clean_performance = self.df[self.df['attack'] == 'Clean'].groupby(['method', 'dataset'])['final_accuracy'].mean()
        attack_performance = self.df[self.df['attack'] != 'Clean'].groupby(['method', 'dataset'])['final_accuracy'].mean()
        
        if not clean_performance.empty and not attack_performance.empty:
            robustness_score = (attack_performance / clean_performance * 100).round(1)
            print("Robustness Score (Attack/Clean Performance %):")
            print(robustness_score.unstack())
        
        # Create comprehensive visualization
        self._create_comprehensive_plots()
        
        return overall_summary
    
    def _create_comprehensive_plots(self):
        """Create comprehensive visualization plots"""
        fig = plt.figure(figsize=(20, 16))
        
        # 1. Method performance heatmap
        plt.subplot(3, 3, 1)
        pivot_data = self.df.pivot_table(values='final_accuracy', index='method', columns='dataset', aggfunc='mean')
        sns.heatmap(pivot_data, annot=True, fmt='.1f', cmap='RdYlGn', vmin=0, vmax=100)
        plt.title('Method Performance by Dataset')
        
        # 2. Attack impact heatmap
        plt.subplot(3, 3, 2)
        attack_pivot = self.df.pivot_table(values='final_accuracy', index='attack', columns='dataset', aggfunc='mean')
        sns.heatmap(attack_pivot, annot=True, fmt='.1f', cmap='RdYlGn', vmin=0, vmax=100)
        plt.title('Attack Impact by Dataset')
        
        # 3. Method robustness comparison
        plt.subplot(3, 3, 3)
        clean_data = self.df[self.df['attack'] == 'Clean']
        attack_data = self.df[self.df['attack'] != 'Clean']
        
        if not clean_data.empty and not attack_data.empty:
            clean_perf = clean_data.groupby(['method', 'dataset'])['final_accuracy'].mean()
            attack_perf = attack_data.groupby(['method', 'dataset'])['final_accuracy'].mean()
            robustness = (attack_perf / clean_perf * 100).reset_index()
            
            for dataset in self.df['dataset'].unique():
                dataset_rob = robustness[robustness['dataset'] == dataset]
                if not dataset_rob.empty:
                    plt.plot(dataset_rob['method'], dataset_rob['final_accuracy'], 
                           marker='o', label=dataset, linewidth=2)
            
            plt.axhline(y=100, color='red', linestyle='--', alpha=0.7, label='Perfect Robustness')
            plt.title('Robustness Score by Method')
            plt.ylabel('Attack/Clean Performance (%)')
            plt.legend()
            plt.xticks(rotation=45)
        
        # 4-6. Dataset-specific performance
        for i, dataset in enumerate(self.df['dataset'].unique()):
            plt.subplot(3, 3, 4 + i)
            dataset_data = self.df[self.df['dataset'] == dataset]
            
            # Box plot of performance by method and attack
            sns.boxplot(data=dataset_data, x='method', y='final_accuracy', hue='attack')
            plt.title(f'{dataset} Performance')
            plt.xticks(rotation=45)
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        
        # 7. Overall method ranking
        plt.subplot(3, 3, 7)
        method_ranking = self.df.groupby('method')['final_accuracy'].mean().sort_values(ascending=True)
        method_ranking.plot(kind='barh')
        plt.title('Overall Method Ranking')
        plt.xlabel('Average Accuracy (%)')
        
        # 8. Attack severity ranking
        plt.subplot(3, 3, 8)
        attack_severity = self.df.groupby('attack')['final_accuracy'].mean().sort_values(ascending=True)
        attack_severity.plot(kind='barh', color='red', alpha=0.7)
        plt.title('Attack Severity Ranking')
        plt.xlabel('Average Accuracy (%)')
        
        # 9. Performance distribution
        plt.subplot(3, 3, 9)
        for method in self.df['method'].unique():
            method_data = self.df[self.df['method'] == method]['final_accuracy']
            plt.hist(method_data, alpha=0.7, label=method, bins=15)
        plt.title('Performance Distribution')
        plt.xlabel('Accuracy (%)')
        plt.ylabel('Frequency')
        plt.legend()
        
        plt.tight_layout()
        
        # Save plot
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"enhanced_fl_results_{timestamp}.png"
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"\nðŸ“Š Comprehensive results saved as: {filename}")
        plt.show()
        
    def generate_research_insights(self):
        """Generate actionable research insights"""
        print("\nðŸ’¡ RESEARCH INSIGHTS & RECOMMENDATIONS")
        print("=" * 60)
        
        if self.df.empty:
            return {}
        
        insights = {}
        
        # 1. Best overall method
        overall_perf = self.df.groupby('method')['final_accuracy'].mean()
        if not overall_perf.empty:
            best_method = overall_perf.idxmax()
            insights['best_overall'] = best_method
            print(f"ðŸ† Best Overall Method: {best_method} ({overall_perf.max():.1f}% avg)")
        
        # 2. Most robust method
        clean_perf = self.df[self.df['attack'] == 'Clean'].groupby('method')['final_accuracy'].mean()
        attack_perf = self.df[self.df['attack'] != 'Clean'].groupby('method')['final_accuracy'].mean()
        
        if not clean_perf.empty and not attack_perf.empty:
            robustness_scores = attack_perf / clean_perf
            most_robust = robustness_scores.idxmax()
            insights['most_robust'] = most_robust
            print(f"ðŸ›¡ï¸ Most Robust Method: {most_robust} ({robustness_scores.max()*100:.1f}% robustness)")
        
        # 3. Dataset-specific recommendations
        print(f"\nðŸ“‹ Dataset-Specific Recommendations:")
        for dataset in self.df['dataset'].unique():
            dataset_perf = self.df[self.df['dataset'] == dataset].groupby('method')['final_accuracy'].mean()
            if not dataset_perf.empty:
                best_for_dataset = dataset_perf.idxmax()
                print(f"   {dataset}: {best_for_dataset} ({dataset_perf.max():.1f}%)")
                insights[f'best_for_{dataset}'] = best_for_dataset
        
        # 4. Attack-specific insights
        print(f"\nðŸŽ¯ Attack-Specific Insights:")
        for attack in self.df['attack'].unique():
            if attack != 'Clean':
                attack_data = self.df[self.df['attack'] == attack]
                best_vs_attack = attack_data.groupby('method')['final_accuracy'].mean()
                if not best_vs_attack.empty:
                    best_method = best_vs_attack.idxmax()
                    print(f"   vs {attack}: {best_method} ({best_vs_attack.max():.1f}%)")
                    insights[f'best_vs_{attack}'] = best_method
        
        # 5. Performance gaps
        print(f"\nðŸ“Š Key Performance Gaps:")
        method_std = self.df.groupby('method')['final_accuracy'].std()
        most_consistent = method_std.idxmin()
        most_variable = method_std.idxmax()
        
        print(f"   Most Consistent: {most_consistent} (Â±{method_std.min():.1f}%)")
        print(f"   Most Variable: {most_variable} (Â±{method_std.max():.1f}%)")
        
        insights['most_consistent'] = most_consistent
        insights['most_variable'] = most_variable
        
        return insights

def main():
    """Enhanced main execution"""
    print("ðŸš€ ENHANCED FL ROBUSTNESS EVALUATION")
    print("ðŸŽ¯ Multi-Dataset Analysis of FL Methods vs Adversarial Attacks")
    print("=" * 70)
    
    # Create enhanced config
    config = EnhancedExperimentConfig()
    
    # Run enhanced experiments
    evaluator = EnhancedRobustnessEvaluator(config)
    results = evaluator.run_all_experiments()
    
    # Save results
    if config.save_results:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        results_file = f"enhanced_fl_results_{timestamp}.json"
        
        json_results = {}
        for key, experiments in results.items():
            json_results[key] = []
            for exp in experiments:
                json_exp = exp.copy()
                for k, v in json_exp.items():
                    if isinstance(v, np.ndarray):
                        json_exp[k] = v.tolist()
                json_results[key].append(json_exp)
        
        with open(results_file, 'w') as f:
            json.dump(json_results, f, indent=2)
        print(f"ðŸ’¾ Results saved to: {results_file}")
    
    # Comprehensive analysis
    analyzer = EnhancedResultsAnalyzer(results)
    summary = analyzer.comprehensive_analysis()
    insights = analyzer.generate_research_insights()
    
    print(f"\nðŸŽ‰ Enhanced evaluation completed!")
    print(f"ðŸ“Š {len(results)} configurations tested across {len(config.datasets)} datasets")
    print(f"ðŸ”¬ {len(config.fl_methods)} methods evaluated vs {sum(len(s) for s in config.attack_scenarios.values())} attack scenarios")
    print(f"âš¡ Optimized for 5-7 hour execution window")
    
    return results, analyzer, insights

if __name__ == "__main__":
    results, analyzer, insights = main()
```
